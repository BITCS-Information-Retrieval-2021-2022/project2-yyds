```python
import pymongo

client = pymongo.MongoClient()
db = client['semantic']
col = db["author_col"]
```

```python
for doc in col.aggregate([{'$group': {'_id': {'name':'$name'}, 'count': {'$sum': 1}}},
                         {'$match':{'count':{'$ne':1}}}
                         ], allowDiskUse=True):
    print(doc)
    name = doc['_id']['name']
    result_same = col.find({'name': name})
    list_same = []
    for x in result_same:
        list_same.append(x)
    length = len(list_same)
    print(length)
    list_id = []
    list_papers = []
    list_sameId = []
    # list_real = []
    papercount = 0
    if length > 0:
        for list1 in list_same:
            list_paper1 = []
            if "papers" in list1:
                for x in list1['papers']:
                    list_paper1.append(x)
            for list2 in list_same:
                list_paper2 = []
                if list1['_id'] != list2['_id']:
                    if "papers" in list2:
                        for x in list2['papers']:
                            list_paper2.append(x)
                list0 = list(set(list_paper1) & set(list_paper2))
                if len(list0) > 0:
                    if list_paper1:
                        list_papers.extend(list_paper1)
                    if list_paper2:
                        list_papers.extend(list_paper2)
                    list_id.append(list1['authorId'])
                    list_id.append(list2['authorId'])
        # print(len(list_papers))
        list_papers = list(set(list_papers))
        if len(list_id) > 0:
            list_sameId = list(set(list_id))
            papercount = len(list_papers)
            authorId = list_sameId[0]
            print(authorId)
            list_real = col.find_one({'authorId':authorId})
            list_real['paperCount'] = papercount
            list_real['papers'] = list_papers
            for authorId in list_sameId:
                # print(authorId)
                col.remove({'authorId': authorId})
            print(list_real)
            col.insert_one(list_real)
```
